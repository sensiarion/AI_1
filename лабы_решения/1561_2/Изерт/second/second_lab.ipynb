```python
df['Fare'] = df['Fare'].astype('str')
df['Cabin'] = df['Cabin'].astype('str')
df.sort_values(by='Pclass', ascending=False).head()
```

```python
df = df.drop(['Embarked'], axis=1)
df.apply(np.max)
```

```python
d = {'male': True, 'female': False}
df['isMale'] = df['Sex'].map(d)
df.head()
```

```python
def f(x):
    try:
        return int(x)
    except:
        return 0
df['Age'] = df['Age'].apply(f)
```
